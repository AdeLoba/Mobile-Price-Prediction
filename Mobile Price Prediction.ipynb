Building and comparing Models

```python
X = df.drop('price_range', axis = 1)
```

```python
Y = df['price_range']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=101)
```

```python
#Using Linear Regression Model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

```python
lm.score(X_test, Y_test)
```

```python
#Using Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, Y_train)
```

```python
dtree.score(X_test, Y_test)
```

```python
#Using Random Forest Model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=250)
```

```python
rfc.fit(X_train, Y_train)
```

```python
rfc.score(X_test, Y_test)
```

Linear Regression and Random Forest are the best of the Model

```python
y_rand = rfc.predict(X_test)
```

```python
plt.scatter(Y_test, y_rand)
```

```python
plt.plot(Y_test, y_rand)
```

```python
plt.figure(figsize=(12, 6))


ax1 = sns.distplot(df['price_range'], hist=False, color="r", label="Actual Value")
sns.distplot(y_rand, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of phones')

plt.show()
plt.close()
```
